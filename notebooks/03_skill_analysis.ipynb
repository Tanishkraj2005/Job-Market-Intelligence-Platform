```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:/Skill Graph/Data/processed/skill_matrix.csv")
df.head()
```

```python
skills = [
    "python","sql","excel","tableau","power bi",
    "pandas","numpy","r","spark","aws",
    "machine learning","statistics","data visualization","hadoop"
]
```

```python

skill_counts = df[skills].sum().sort_values(ascending=False)
skill_counts
```

```python

skill_df = skill_counts.reset_index()
skill_df.columns = ["Skill", "Job Count"]

skill_df.head()
```

```python
top_skills = skill_df.head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x="Job Count",
    y="Skill",
    data=top_skills
)

plt.title("Top 10 Most Demanded Data Skills")
plt.xlabel("Number of Job Postings")
plt.ylabel("Skill")

plt.savefig("D:/Skill Graph/visualization/top_skill_demand.png")

plt.show()
```
